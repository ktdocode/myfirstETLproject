SIMPLE ETL PROJECT

1. Import Data

```python
import pandas as pd
from datetime import datetime
```

2. Extract data

```python
def extract_from_csv(file_to_process):
    data = pd.read_csv(file_to_process)
    return data
```

```python
def extract():
    df = pd.DataFrame()
    df = df.append(extract_from_csv("points_table.csv") , ignore_index = True)
    return df
df = extract()
df
```

3. Transform data

```python
def transform(df):
    # Merge columns
    df["W-D-L"] = df['W'].astype(str)+"-"+ df["D"].astype(str)+"-" + df["L"].astype(str)
    # Remove columns
    df = df.drop(columns = ["W" , "D" , "L"])
    # Change position of columns
    df_new_cols =  ["Pos" , "Team" , "Pld" , "GF" , "W-D-L" , "GA" , "GD" , "Pts"]
    df = df[df_new_cols]
    df = df.reindex(columns = df_new_cols)
    return df
df = transform(df)
df
```

4. Load data

```python
targetfile = "epl_standing_2122.csv"
def load(targetfile , data_to_load):
    data_to_load.to_csv(targetfile)
# Call function to load data
load(targetfile , df)
```

5. Logging Function

```python
def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Year-Monthname-Day-Hour-Minute-Second
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("logfile.txt","a") as f:
        f.write(timestamp + ',' + message + '\n')
```

RUNNING ETL PROCESS

```python
log("ETL Job Started")
```

Extract

```python
log("Extract phase Started")
extracted_data = extract()
# Print Big Six
extracted_data[:6]
```

```python
log("Extract phase Ended")
```

Transform

```python
log("Transform phase Started")
```

```python
transformed_data = transform(extracted_data)
transformed_data[:6]
```

```python
log("Transform phase Ended")
```

Load

```python
log("Load phase Started")
```

```python
load(targetfile , transformed_data)
```

```python
log("Load phase Ended")
```
